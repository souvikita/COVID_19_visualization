### Merging the datasets so that we can have confirmed, deaths and recovered in one DataFrame

```python
dates = confirmed_df.columns[4:]
confirmed_df_long = confirmed_df.melt(                           #Use the melt() function to un-pivot the database.
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], # convert the wide format to a long format data.
    value_vars=dates, 
    var_name='Date', 
    value_name='Confirmed'
)
deaths_df_long = deaths_df.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
    value_vars=dates, 
    var_name='Date', 
    value_name='Deaths'
)
recovered_df_long = recovered_df.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
    value_vars=dates, 
    var_name='Date', 
    value_name='Recovered'
)
```

```python
confirmed_df_long #Note the difference in this dataframe compared to the earlier original.
```

```python
recovered_df_long
```

```python
recovered_df_long = recovered_df_long[recovered_df_long['Country/Region']!='Canada'] ## Canada recovered data is counted by Country-wise rather than Province/State-wise
```

```python
# Merging confirmed_df_long and deaths_df_long
full_table = confirmed_df_long.merge(
  right=deaths_df_long, 
  how='left',
  on=['Province/State', 'Country/Region', 'Date', 'Lat', 'Long']
)
# Merging full_table and recovered_df_long
full_table = full_table.merge(
  right=recovered_df_long, 
  how='left',
  on=['Province/State', 'Country/Region', 'Date', 'Lat', 'Long']
)
```

```python
full_table # see how beautiful.....
```

### We need to perform data cleaning-
* Lot of NaNs in the data above.
* Three cruise ships are also there. We need to analyse them separately. 
* Converting Date from string to datetime for the ease of access and analysis.

```python
# checking for NaNs
full_table.isna().sum()
```

A lot of NaNs are present in the _Province_ column and a few thousand in the _Recovered_ column

```python
full_table['Recovered'] = full_table['Recovered'].fillna(0) # replacing the NaNs in the data by 0
```

```python
# Cruise ships
ship_rows = full_table['Province/State'].str.contains('Grand Princess') | full_table['Province/State'].str.contains('Diamond Princess') | full_table['Country/Region'].str.contains('Diamond Princess') | full_table['Country/Region'].str.contains('MS Zaandam')
```

```python
ship_rows # basically we checked if the three ships namely Grand Princess
          # Diamond Princess and MS Zaandam are present either in the Country/Region column 
          # or in the Province/State column
```

Apparently it is quite simple to get rid of the ship data from the `full_table` by simply using the `~` symbol. How fascinating????!!!!

```python
full_table = full_table[~(ship_rows)] # like this.. Oh Lots of new and cool stuff to learn. I feel overwhelmed at times.
```

Note the difference in the **rows** from the previous version of `full_table`. It has reduced due to eliminating the ships. 

### Agrregating the data set 
Also adding another column called _**Active**_ by 
`active = confirmed — deaths — recovered`

```python
# Active Case = confirmed - deaths - recovered
full_table['Active'] = full_table['Confirmed'] - full_table['Deaths'] - full_table['Recovered'] 
```

Yes that's it. easy peasy!!!! You can ignore the warnings above but it is good to know that `iloc` and `loc` is a better choice perhaps

Now, let’s aggregate data into `Country/Region` wise and group them by `Date` and `Country/Region`. Because that will help in the end to access the data by country.

```python
full_grouped = full_table.groupby(['Date', 'Country/Region'])['Confirmed', 'Deaths', 'Recovered', 'Active'].sum().reset_index()
```

* `sum()` is to get the total count of ‘Confirmed’, ‘Deaths’, ‘Recovered’, ‘Active’ for the given **Date** and **Country/Region**.
* `reset_index()` is used to change the index to a default index here it is `Date`.

Now `full_grouped` looks like the following: 

```python
full_grouped
```

Now we might want to do something similar to what most databases including [Worldometers](https://www.worldometers.info/coronavirus/) and the amazing Indian database [covid19india.org](https://www.covid19india.org/) does. It is to add new columns such as **New cases, New deaths**, and **New recovered** to the above database. For that, we would have to deduct the corresponding accumulative data on the previous day.

#### The suggestion to do this came from one of the developers involved in the Covid19India.org database named Priya Jaiswal. Kudos to her

```python
# new cases 

temp = full_grouped.groupby(['Country/Region', 'Date', ])['Confirmed', 'Deaths', 'Recovered']
temp = temp.sum().diff().reset_index()
mask = temp['Country/Region'] != temp['Country/Region'].shift(1)
temp.loc[mask, 'Confirmed'] = np.nan
temp.loc[mask, 'Deaths'] = np.nan
temp.loc[mask, 'Recovered'] = np.nan
# renaming columns
temp.columns = ['Country/Region', 'Date', 'New cases', 'New deaths', 'New recovered']
# merging new values
full_grouped = pd.merge(full_grouped, temp, on=['Country/Region', 'Date'])
# filling na with 0
full_grouped = full_grouped.fillna(0)
# fixing data types
cols = ['New cases', 'New deaths', 'New recovered']
full_grouped[cols] = full_grouped[cols].astype('int')
# 
full_grouped['New cases'] = full_grouped['New cases'].apply(lambda x: 0 if x<0 else x)
```

Finally the cleaned data set is here: `full_grouped`. Notice the newer columns on the right named as `New cases`, `New Deaths` and `New recovered`

We analyse the spread trends by the so-called fancy [Altair visualization](https://altair-viz.github.io/user_guide/customization.html). The size of the circle in time will show the number of cases and how big the cluster has grown as it evolves in time for the different countries. 

Let's create circle chart for the daily **New cases**

```python
alt.Chart(selected_countries).mark_circle().encode(
    x='monthdate(Date):O',
    y='Country/Region',
    color='Country/Region',
    size=alt.Size('New cases:Q',
        scale=alt.Scale(range=[0, 5000]),
        legend=alt.Legend(title='Daily new cases')
    ) 
).properties(
    width=800,
    height=300
) # taking the selected_countries as given in the row.
```
